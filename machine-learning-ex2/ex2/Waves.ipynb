```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy import dot,sum,log
from numpy.linalg import inv
import scipy.optimize as opt
nums = np.arange(-10, 10, step=1)

fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, np.tanh(nums), 'r')  
```

```python
fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, np.sin(nums), 'r') 
```

```python
fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, np.exp(nums), 'r') 
```

```python
fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, np.exp(-nums), 'r') 
```

```python
fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, 1/(1+np.exp(nums)), 'r') 
```

```python
fig, ax = plt.subplots(figsize=(3,3))  
ax.plot(nums, 1/(1-np.exp(nums)), 'r') 
```
